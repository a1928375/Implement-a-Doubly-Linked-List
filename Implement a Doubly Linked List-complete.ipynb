```python
class DoublyLinkedListNode(object):
    
    def __init__(self, value, nextnode = None, prevnode = None):
        
        self.value = value
        self.nextnode = nextnode
        self.prevnode = prevnode
        
class DoublyLinkedList(object):
    
    def __init__(self, size=0):
        
        self.head = None
        self.size = size
        
    def getSize(self):
        
        return self.size
        
    def addFrontNode(self, value):
        
        newnode = DoublyLinkedListNode(value,self.head)
        
        if self.head is None:

            self.head = newnode
            self.size = self.size+1
            
            return
        
        self.head.prevnode = newnode
        
        self.head = newnode
        
        self.size = self.size+1
        
    def addRearNode(self, value):
        
        newnode = DoublyLinkedListNode(value)
        
        if self.head is None:
                       
            self.head = newnode
            
            return
        
        n = self.head
        
        while n.nextnode != None:
            
            n = n.nextnode
        
        n.nextnode = newnode
        
        newnode.prevnode = n
        
        self.size = self.size+1
        
    def deleteFrontNode(self):
        
        if self.head == None:
            
            print("The list has no element to delete")
            
            return 
        
        if self.head.nextnode == None:
            
            self.head = None
            
            return
        
        self.head = self.head.nextnode
        
        self.head.prevnode = None
        
        self.size = self.size-1
        
    def deleteRearNode(self):
    
        if self.head == None:
            
            print("The list has no element to delete")
            
            return 
        
        if self.head.nextnode == None:
            
            self.head = None
            
            return
        
        n = self.head
        
        while n.nextnode != None:
            
            n = n.nextnode
            
        n.prevnode.nextnode = None      #有nextnode & prevnode，要善加利用
        #n = None                       此句只讓 n=None，並沒有改變任何DoublyLinkedList的sequence
        
        self.size = self.size-1


    def printNode(self):
        
        curr = self.head
        
        while curr:
            
            print (curr.value)
            
            curr = curr.nextnode


myList = DoublyLinkedList()

myList.addFrontNode(1)
myList.addFrontNode(2)
myList.addFrontNode(3)

myList.addRearNode(0)

myList.deleteFrontNode()

myList.deleteRearNode()

myList.printNode()

print (myList.getSize())
```
